```python
def get_id2info_dicts(id_mean_df):
    """
    Get mapping from tag ids to tag type and from tag ids to tag description 
    """
    id2desc_dict = dict()
    id2type1_dict = dict()
    id2type2_dict = dict()
    for index, row in id_mean_df.iterrows():
    #     print(type(row['tag_code']), row['tag_type'], row['tag_desc'])
        id2type1_dict[str(row['tag_code'])] = row['tag_type1']
        id2type2_dict[str(row['tag_code'])] = row['tag_type2']
        id2desc_dict[str(row['tag_code'])] = row['tag_desc']
    return id2desc_dict, id2type1_dict, id2type2_dict
```

```python
def create_app_install_usage_map(app_install_ids, app_usage_ids):
    """
    create mapping from app install ids to app usage ids, and vice versa.
    """
    install2usage_dict = dict()
    usage2install_dict = dict()
    for install_id, usage_id in zip(app_install_ids, app_usage_ids):
        install_id_str = str(install_id)
        usage_id_str = str(usage_id)
        install2usage_dict[install_id_str] = usage_id_str
        usage2install_dict[usage_id_str] = install_id_str
    return install2usage_dict, usage2install_dict
```

```python
def get_corid2info_dict(orgid2info_dict, orgid2corid_dict):
    """
    """
    corid2info_dict = dict()
    for key, info in orgid2info_dict.items():
        corid2info_dict[orgid2corid_dict[key]] = info
    return corid2info_dict
```
