```python
import pandas as pd
import numpy as np
```

```python
def get_tag_id2info_dicts(id_mean_df):
    """
    Get mapping from tag ids to tag type and from tag ids to tag description
    """
    id2desc_dict = dict()
    id2type1_dict = dict()
    id2type2_dict = dict()
    for index, row in id_mean_df.iterrows():
        #     print(type(row['tag_code']), row['tag_type'], row['tag_desc'])
        id2desc_dict[str(row['tag_code'])] = row['tag_desc']
        id2type1_dict[str(row['tag_code'])] = row['tag_type1']
        id2type2_dict[str(row['tag_code'])] = row['tag_type2']
    return id2desc_dict, id2type1_dict, id2type2_dict
```

```python
id_mean_v4_df = pd.read_csv('../../../data/cell_manager/ID_meaning_v5.csv')
id2desc_dict, id2type1_dict, id2type2_dict = get_tag_id2info_dicts(id_mean_v4_df)
```

```python
id2desc_dict
```

```python
def show_col_value_unique(df):
    for col in df.columns:
#         print("col:", col)
        print("col:", col, df[col].unique())
```
